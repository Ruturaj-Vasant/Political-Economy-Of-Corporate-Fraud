```python
import os
import pandas as pd
os.getcwd()
```

```python
file_path = os.path.join("/Users/ruturaj_vasant/Desktop/PersonalProjects/Political-Economy-Of-Corporate-Fraud/data/XRX/extracted/XRX_2020_SCT.csv")
with open(file_path,"rb") as f:
    content = f.read()
    print(content)
```

```python
df = pd.read_csv(file_path)
print(df.info())
# df.head
```

```python
columns_with_nan = df.columns[df.columns.str.contains('nan')]
# print(columns_with_nan)
col_new_names ={}
for col in columns_with_nan:
    new_name = col.replace('nan', "")
    col_new_names[col] = new_name

df = df.rename(columns=col_new_names)

df.head
```

```python
df = df.astype(str)
df = df.replace('nan', '', regex=True)
df = df.dropna()
print(df.info())
df.head
```

```python
df.dropna(how="all")
df.head
```

```python
df.to_csv("cleaned_csv.csv")
```
