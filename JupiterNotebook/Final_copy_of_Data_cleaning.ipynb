###Basic text cleaning

````python
def clean_text_basic(raw: str) -> str:
    if raw is None:
        return ""

    s = str(raw)
    s = s.replace("\u00a0", " ")                     # non-breaking space
    s = re.sub(r"(\w)-\s+(\w)", r"\1\2", s)         # fix hyphen splits
    s = s.replace("/", " ")                         # slash spacing
    s = re.sub(r"\s+", " ", s).strip()              # collapse whitespace
    s = re.sub(r"\s*\([^)]*\)", "", s)              # footnotes
    s = s.replace("$", "")
    s = s.replace(",", "")
    s = re.sub(r"\s+", " ", s).strip().lower()

    if not s:
        return ""

    return s

def clean_basic_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    df2 = df.copy()
    for col in df2.columns:
        df2[col] = df2[col].apply(
            lambda x: clean_text_basic(x) if isinstance(x, str) else x
        )
    return df2
````

###Normalize The Headers

````python
import re
import pandas as pd

def normalize_semantic_header(raw: str) -> str:
    if raw is None:
        return ""

    s = str(raw).lower().strip()

    # semantic mapping (SCT-focused)
    if "name" in s and "position" in s:
        return "name_position"
    if "fiscal" in s or "year" in s:
        return "year"
    if "salary" in s:
        return "salary"
    if "bonus" in s:
        return "bonus"
    if "other annual" in s:
        return "other_annual_comp"
    if "stock" in s and "award" in s:
        return "stock_awards"
    if "option" in s and "award" in s:
        return "option_awards"
    if "options" in s or "option" in s:
        return "option"
    if "incentive" in s:
        return "non_equity_incentive"
    if "pension" in s or "deferred compensation" in s:
        return "pension_value"
    if "all other" in s:
        return "all_other_comp"
    if "total" in s:
        return "total"

    # fallback snake_case
    s = re.sub(r"[^a-z0-9]+", "_", s)
    return s.strip("_")

def normalize_headers_only(df: pd.DataFrame) -> pd.DataFrame:
    df2 = df.copy()
    df2.columns = [normalize_semantic_header(c) for c in df2.columns]
    return df2

# apply to all dataframes in your dict
for df_name, df in csv_dataframes.items():
    csv_dataframes[df_name] = normalize_headers_only(df)
    print(f"Processed: {df_name}")
````

###Dejunk ROws below header

````python
import re
import pandas as pd
import numpy as np

YEAR_RE = re.compile(r"(19|20)\d{2}")

HEADER_JUNK_RE = re.compile(
    r"(name|principal|position|year|fiscal|salary|bonus|stock|option|incentive|pension|all other|total)",
    re.IGNORECASE,
)

def drop_pre_data_junk_rows(df: pd.DataFrame, year_col: str = "year") -> pd.DataFrame:
    df2 = df.copy()

    if year_col not in df2.columns:
        return df2

    # Find first row that contains a year anywhere (handles "2006 2005" too)
    year_has_year = df2[year_col].apply(lambda x: bool(YEAR_RE.search(str(x))) if not pd.isna(x) else False)
    if not year_has_year.any():
        return df2

    first_data_idx = year_has_year[year_has_year].index.min()

    # Only inspect rows from top to first_data_idx-1
    top = df2.loc[:first_data_idx - 1].copy()

    def row_is_junk(r) -> bool:
        joined = " ".join("" if pd.isna(v) else str(v) for v in r.values)

        # If it contains header-like tokens, it's likely a leftover header row
        if HEADER_JUNK_RE.search(joined):
            return True

        # Or if it contains no digits at all, it’s usually just structural/header fluff
        if not re.search(r"\d", joined):
            return True

        return False

    junk_mask = top.apply(row_is_junk, axis=1)

    # Keep only non-junk rows in the top section, then append from first_data_idx onward
    cleaned_top = top.loc[~junk_mask]
    cleaned = pd.concat([cleaned_top, df2.loc[first_data_idx:]], axis=0).reset_index(drop=True)

    return cleaned

# Apply to all tables
for df_name, df in csv_dataframes.items():
    csv_dataframes[df_name] = drop_pre_data_junk_rows(df, year_col="year")
    print(f"Processed: {df_name}")
````

### Expand Columns

````python
import re
import numpy as np
import pandas as pd

# === Config (normalized headers) ===
SCT_NUMERIC_COLS = {
    "salary",
    "bonus",
    "option",
    "other_annual_comp",
    "stock_awards",
    "option_awards",
    "non_equity_incentive",
    "pension_value",
    "all_other_comp",
    "total",
}

# === Regex / parsing ===
YEAR_FIND_RE = re.compile(r"(?:19|20)\d{2}")
YEAR_FULL_RE = re.compile(r"^(?:19|20)\d{2}$")
SPLIT_RE = re.compile(r"\s+")
PLACEHOLDERS = {"—", "–", "-", ""}

def _split_tokens_keep_slots(x):
    """Split by whitespace/newlines; keep placeholder tokens as slots."""
    if pd.isna(x):
        return []
    s = str(x).strip()
    if s == "":
        return []
    s = s.replace("–", "—")
    return [t for t in SPLIT_RE.split(s) if t is not None and t != ""]

def _to_num_or_zero(tok):
    """Placeholder -> 0.0, numeric -> float, unparseable -> NaN."""
    if tok is None:
        return 0.0
    s = str(tok).strip()
    if s in PLACEHOLDERS or s == "—":
        return 0.0
    s = s.replace(",", "").replace("$", "")
    s = s.replace("(", "-").replace(")", "")
    s = re.sub(r"[^\d\.\-]", "", s)
    if s in {"", "-", "."}:
        return 0.0
    try:
        return float(s)
    except:
        return np.nan

def _find_driver_year_col(df: pd.DataFrame, year_name: str = "year") -> int | None:
    """
    Find the first 'year' column (by position) that contains a multi-year cell (2+ years).
    Works with duplicate column names; does not collapse columns.
    """
    year_col_idxs = [i for i, c in enumerate(df.columns) if c == year_name]
    if not year_col_idxs:
        return None

    for col_i in year_col_idxs:
        for v in df.iloc[:, col_i].tolist():
            if pd.isna(v):
                continue
            if len(YEAR_FIND_RE.findall(str(v))) >= 2:
                return col_i
    return None

def expand_stacked_rows_if_needed(df: pd.DataFrame, year_name: str = "year") -> pd.DataFrame:
    """
    Single entrypoint:
      1) Detect if table needs expansion (multi-year stacked cells).
      2) If not needed, return df unchanged.
      3) If needed, expand rows with stacked years and align numeric cols by slot,
         treating dash/blank placeholders as 0.0.
    Does NOT collapse duplicate columns; uses column positions (iloc).
    """
    driver_year_col = _find_driver_year_col(df, year_name=year_name)
    if driver_year_col is None:
        return df  # no expansion needed

    col_names = list(df.columns)  # keep duplicates and order
    out_rows = []

    for r_i in range(len(df)):
        row_cells = df.iloc[r_i, :].tolist()

        # derive year list from the driver year col
        year_cell = row_cells[driver_year_col]
        driver_years = YEAR_FIND_RE.findall("" if pd.isna(year_cell) else str(year_cell))

        # if this row isn't multi-year, keep as-is
        if len(driver_years) < 2:
            out_rows.append(row_cells)
            continue

        n = len(driver_years)

        # pre-tokenize numeric/year columns by column position
        token_lists = [None] * len(col_names)
        for c_i, c_name in enumerate(col_names):
            if c_name == year_name:
                toks = [t for t in _split_tokens_keep_slots(row_cells[c_i]) if YEAR_FULL_RE.match(t)]
                token_lists[c_i] = toks
            elif c_name in SCT_NUMERIC_COLS:
                token_lists[c_i] = _split_tokens_keep_slots(row_cells[c_i])
            else:
                token_lists[c_i] = None  # replicate text cols

        # emit expanded rows
        for k in range(n):
            new_row = []
            for c_i, c_name in enumerate(col_names):
                cell = row_cells[c_i]

                # Year columns: use kth year if available else fallback to driver list
                if c_name == year_name:
                    toks = token_lists[c_i] or []
                    if len(toks) >= n:
                        new_row.append(int(toks[k]))
                    elif len(toks) == 1 and YEAR_FULL_RE.match(toks[0]):
                        new_row.append(int(toks[0]))
                    else:
                        new_row.append(int(driver_years[k]))
                    continue

                # Numeric cols: zip by slot, pad with 0.0
                if c_name in SCT_NUMERIC_COLS:
                    toks = token_lists[c_i] or []
                    if len(toks) >= n:
                        new_row.append(_to_num_or_zero(toks[k]))
                    elif len(toks) == 1:
                        new_row.append(_to_num_or_zero(toks[0]))
                    elif len(toks) == 0:
                        new_row.append(0.0)
                    else:
                        new_row.append(_to_num_or_zero(toks[k]) if k < len(toks) else 0.0)
                    continue

                # Text/other cols: replicate
                new_row.append(cell)

            out_rows.append(new_row)

    return pd.DataFrame(out_rows, columns=col_names).reset_index(drop=True)

# === Apply to all tables in your dict ===
for df_name, df in csv_dataframes.items():
    before = df.shape
    csv_dataframes[df_name] = expand_stacked_rows_if_needed(df, year_name="year")
    after = csv_dataframes[df_name].shape
    print(f"Processed: {df_name} | {before} -> {after}")
````

## Identify Executive Blocks

### Subtask:
Implement a function to iterate through each DataFrame, identify executive blocks based on the year pattern (non-increasing years within a block, increasing year indicates a new block), and assign a unique 'exec_block_id' to each block. Rows with blank years should belong to the current block.


**Reasoning**:
The subtask requires implementing a function to identify executive blocks and then applying it to all DataFrames. First, I will define the `identify_executive_blocks` function as specified in the instructions, which includes handling year patterns and assigning `exec_block_id`.



````python
import numpy as np

def identify_executive_blocks(df: pd.DataFrame) -> pd.DataFrame:
    df_copy = df.copy()
    df_copy['exec_block_id'] = -1

    # Convert 'year' to numeric for proper comparison
    df_copy['year_numeric'] = pd.to_numeric(df_copy['year'], errors='coerce')

    current_exec_block_id = 0
    previous_year = None

    for i in range(len(df_copy)):
        current_year = df_copy.loc[i, 'year_numeric']
        name_position = df_copy.loc[i, 'name_position']

        # Condition 1: Increasing year indicates a new block
        if pd.notna(current_year) and previous_year is not None and current_year > previous_year:
            current_exec_block_id += 1
        # Condition 2: Handle blank years and new blocks
        elif pd.isna(current_year) and pd.notna(name_position) and name_position != '':
            # If it's the very first row, start with block 0
            if i == 0:
                current_exec_block_id = 0
            else:
                # Check if previous row had a non-NaN year and empty name_position
                prev_name_position = df_copy.loc[i-1, 'name_position']
                prev_year_numeric = df_copy.loc[i-1, 'year_numeric']

                if pd.notna(prev_year_numeric) and (pd.isna(prev_name_position) or prev_name_position == '') : # Adjusted logic as per prompt
                    current_exec_block_id += 1

        df_copy.loc[i, 'exec_block_id'] = current_exec_block_id

        if pd.notna(current_year):
            previous_year = current_year

    # Fill forward any remaining -1 values (e.g., if first rows had NaN years and no name_position to trigger a block change)
    df_copy['exec_block_id'] = df_copy['exec_block_id'].replace(-1, np.nan).ffill().fillna(0).astype(int)

    # Drop the temporary column
    df_copy = df_copy.drop(columns=['year_numeric'])
    return df_copy

print("Defined identify_executive_blocks function.")
````

**Reasoning**:
Now that the `identify_executive_blocks` function is defined, I will apply it to each DataFrame in `csv_dataframes` as specified in the instructions, then display a sample DataFrame to show the new 'exec_block_id' column.



````python
for df_name, df in csv_dataframes.items():
    if 'year' in df.columns and 'name_position' in df.columns:
        csv_dataframes[df_name] = identify_executive_blocks(df)
        print(f"Processed {df_name}: Identified executive blocks.")
    else:
        print(f"Skipping {df_name}: Missing 'year' or 'name_position' column.")

# Display the first few rows of one modified DataFrame to show the new 'exec_block_id'
sample_df_name = next(iter(csv_dataframes))
print(f"\nFirst 10 rows of {sample_df_name} after identifying executive blocks:")
display(csv_dataframes[sample_df_name].head(10))
````

````python
def consolidate_name_position(df: pd.DataFrame) -> pd.DataFrame:
    df_copy = df.copy()
    if 'name_position' not in df_copy.columns or 'exec_block_id' not in df_copy.columns:
        return df_copy

    # Group by exec_block_id and consolidate name_position
    def combine_names(series):
        # Filter out NaN, empty strings, and duplicates, then join
        clean_names = series.dropna().astype(str).apply(lambda x: x.strip()).loc[lambda x: x != ''].unique()
        if len(clean_names) > 0:
            return ' | '.join(clean_names)
        return ""

    consolidated_names = df_copy.groupby('exec_block_id')['name_position'].transform(combine_names)
    df_copy['name_position_consolidated'] = consolidated_names

    return df_copy

# Apply to all dataframes in csv_dataframes
for df_name, df in csv_dataframes.items():
    csv_dataframes[df_name] = consolidate_name_position(df)
    print(f"Processed {df_name}: Consolidated name and position.")

# Display the first few rows of one modified DataFrame to show the new 'name_position_consolidated' column
sample_df_name = next(iter(csv_dataframes))
print(f"\nFirst 10 rows of {sample_df_name} after consolidating name and position:")
display(csv_dataframes[sample_df_name].head(10))
````

````python
import re
import pandas as pd
import numpy as np

def dataframe_to_json_format(df: pd.DataFrame, df_name: str) -> dict:
    # 2. Extract ticker and report_date from df_name
    parts = df_name.split('_')
    ticker = parts[0]
    report_date = parts[1] + '-' + parts[2] + '-' + parts[3]

    # 3. Initialize a list of compensation columns
    compensation_columns = [
        'salary',
        'bonus',
        'stock_awards',
        'option_awards',
        'non_equity_incentive',
        'pension_value',
        'all_other_comp',
        'total',
    ]

    # 4. Create an empty dictionary for the final JSON structure
    json_output = {
        'ticker': ticker,
        'report_date': report_date,
        'executives': []
    }

    # 5. Iterate through the input DataFrame, grouping by executive_name
    for exec_name, exec_df in df.groupby('executive_name'):
        executive_data = {
            'executive_name': exec_name,
            'compensation_data': []
        }

        # 6.b. Iterate through the rows belonging to this executive, grouping by year
        for year, year_df in exec_df.groupby('year'):
            # Assuming there's only one row per executive and year combination for compensation
            # If there are multiple, we'll take the first one or sum, for now taking first.
            row = year_df.iloc[0]

            year_comp_data = {
                'year': int(float(year)) if pd.notna(year) and str(year).strip() != '' else None,
                'position': row.get('position', '')
            }

            # 6.c.iii. Process the compensation columns
            col_counts = {}
            for col in df.columns:
                if col in compensation_columns:
                    original_col_name = col
                    # Handle duplicate column names
                    if original_col_name in col_counts:
                        col_counts[original_col_name] += 1
                        col_key = f"{original_col_name}_{col_counts[original_col_name]}"
                    else:
                        col_counts[original_col_name] = 1
                        col_key = original_col_name

                    # Convert to numeric, handle non-numeric or missing values to 0.0
                    value = row[col]
                    try:
                        year_comp_data[col_key] = float(value) if pd.notna(value) and str(value).strip() != '' else 0.0
                    except ValueError:
                        year_comp_data[col_key] = 0.0

            # 6.c.iv. Identify and collect other_fields
            other_fields = {}
            known_cols = set(compensation_columns + ['executive_name', 'year', 'position', 'exec_block_id', 'name_position', 'name_position_consolidated'])
            for col in df.columns:
                if col not in known_cols and pd.notna(row[col]) and str(row[col]).strip() != '':
                    other_fields[col] = str(row[col])
            if other_fields:
                year_comp_data['other_fields'] = other_fields

            executive_data['compensation_data'].append(year_comp_data)
        json_output['executives'].append(executive_data)

    return json_output

print("Defined `dataframe_to_json_format` function.")
````

````python
import json
import os
import re
import numpy as np
import pandas as pd

def dataframe_to_json_format(df: pd.DataFrame, df_name: str) -> dict:
    # 2. Extract ticker and report_date from df_name
    parts = df_name.split('_')
    ticker = parts[0]
    # Handle cases where df_name might be shorter than expected if it's not a SCT file like AZPN_2023-10-24_SCT
    if len(parts) >= 4:
        report_date = f"{parts[1]}-{parts[2]}-{parts[3]}"
    else:
        # Fallback or error handling if df_name format is unexpected
        report_date = "unknown_date"


    # 3. Initialize a list of compensation columns
    compensation_columns = [
        'salary',
        'bonus',
        'stock_awards',
        'option_awards',
        'non_equity_incentive',
        'pension_value',
        'all_other_comp',
        'total',
    ]

    # 4. Create an empty dictionary for the final JSON structure
    json_output = {
        'ticker': ticker,
        'report_date': report_date,
        'executives': []
    }

    # --- Start of robust year cleaning ---
    df_cleaned = df.copy()

    def parse_year_robustly(year_val):
        if pd.isna(year_val):
            return np.nan
        s_year = str(year_val).strip()
        if s_year == '':
            return np.nan
        # First, try direct conversion if it's already a clean number
        try:
            return int(float(s_year))
        except ValueError:
            pass # Not a simple number, proceed to regex

        # If not a simple number, try to extract a 4-digit year using regex
        year_match = re.search(r'\b(19|20)\d{2}\b', s_year)
        if year_match:
            return int(year_match.group(0))
        return np.nan # If no clear year found, return NaN

    df_cleaned['year'] = df_cleaned['year'].apply(parse_year_robustly)

    # Drop rows where year could not be parsed to a valid number
    # If a year cannot be parsed, that row cannot contribute to year-based compensation data.
    df_cleaned = df_cleaned.dropna(subset=['year']).copy()
    if not df_cleaned.empty: # Only attempt conversion if there are rows left
        df_cleaned['year'] = df_cleaned['year'].astype(int) # Ensure year is integer type
    # --- End of robust year cleaning ---

    # --- Start: Make column names unique to prevent ambiguity errors ---
    cols = df_cleaned.columns.tolist()
    seen = {}
    new_cols = []
    for col in cols:
        if col in seen:
            seen[col] += 1
            new_cols.append(f"{col}_{seen[col]}")
        else:
            seen[col] = 1
            new_cols.append(col)
    df_cleaned.columns = new_cols
    # --- End: Make column names unique ---

    # 5. Iterate through the input DataFrame, grouping by executive_name
    # Only proceed if df_cleaned is not empty after year cleaning
    if not df_cleaned.empty:
        for exec_name, exec_df in df_cleaned.groupby('executive_name'):
            executive_data = {
                'executive_name': exec_name,
                'compensation_data': []
            }

            # 6.b. Iterate through the rows belonging to this executive, grouping by year
            # 'year' here should now be numeric (integer)
            for year, year_df in exec_df.groupby('year'):
                # Take the first row if multiple rows exist for the same executive-year combination.
                row = year_df.iloc[0]

                year_comp_data = {
                    'year': int(year), # Year is now guaranteed to be numeric (int)
                    'position': row.get('position', '')
                }

                # 6.c.iii. Process the compensation columns using the unique column names
                for col_unique_name in df_cleaned.columns:
                    base_col_name = col_unique_name.split('_')[0] # Get base name, e.g., 'salary' from 'salary_2'

                    if base_col_name in compensation_columns:
                        value = row[col_unique_name]
                        try:
                            numeric_value = float(value) if pd.notna(value) and str(value).strip() != '' else 0.0
                        except (ValueError, TypeError):
                            numeric_value = 0.0
                        year_comp_data[col_unique_name] = numeric_value # Use the unique column name as the key

                # 6.c.iv. Identify and collect other_fields
                other_fields = {}
                # Define known original column names (base names) that are NOT compensation columns but are metadata
                base_metadata_cols = {'executive_name', 'year', 'position', 'exec_block_id', 'name_position', 'name_position_consolidated'}

                for col_unique_name in df_cleaned.columns:
                    base_col_name = col_unique_name.split('_')[0]

                    # If the column's base name is NOT a base metadata column
                    # AND its base name is NOT a base compensation column
                    # AND the specific unique column name is not already handled in year_comp_data (like 'salary_2')
                    if base_col_name not in base_metadata_cols and base_col_name not in compensation_columns and col_unique_name not in year_comp_data:
                        val = row[col_unique_name]
                        val_str = str(val).strip() # Convert potentially ambiguous value to string to avoid ValueError
                        if val_str != '':
                            other_fields[col_unique_name] = val_str
                if other_fields:
                    year_comp_data['other_fields'] = other_fields

                executive_data['compensation_data'].append(year_comp_data)
            json_output['executives'].append(executive_data)

    return json_output

# Apply the function to all dataframes in csv_dataframes
json_results = {}
output_dir = "/content/json_output/"
os.makedirs(output_dir, exist_ok=True)

for df_name, df in csv_dataframes.items():
    if 'executive_name' in df.columns and 'year' in df.columns:
        json_data = dataframe_to_json_format(df, df_name)
        json_results[df_name] = json_data

        # Save the JSON object to a file
        output_filename = os.path.join(output_dir, f"{df_name}.json")
        with open(output_filename, 'w') as f:
            json.dump(json_data, f, indent=4)
        print(f"Processed {df_name}: Converted to JSON and saved to {output_filename}")
    else:
        print(f"Skipping {df_name}: Missing 'executive_name' or 'year' column, cannot convert to JSON.")

# Display the first few lines of one of the generated JSON files for verification
if json_results:
    sample_df_name = next(iter(json_results))
    sample_output_filename = os.path.join(output_dir, f"{sample_df_name}.json")
    print(f"\nFirst 10 lines of {sample_output_filename}:")
    with open(sample_output_filename, 'r') as f:
        for _ in range(10):
            line = f.readline()
            if not line: break
            print(line.strip())
````
